
 * Создайте новый фрейм данных на основе набора данных hit_albums, который содержит только исполнителей, у которых в списке есть как минимум два альбома.

````python
two_albums=hit_albums[['Artist','Album']] # Создаем датафрейм со столбцами "Artist", "Album"
two_album_count = two_albums['Artist'].value_counts().to_frame() # Считаем кол-во альбомов у исполнителей и конвертируем в датафрейм
two_album_count.reset_index(inplace= True ) # Добавляем столбец с индексами
two_album_count
````

````python
two_album_count.columns = ['Artist', 'Count_albums'] # Меняем название столбцов
two_album_count
````

````python
artist_album= two_album_count[two_album_count["Count_albums"] >= 2] # Больше 2-х альбомов у исполнителя
artist_two=artist_album['Artist']
artist_two
````
